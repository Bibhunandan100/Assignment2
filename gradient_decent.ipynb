```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([1,2,3])
y=np.array([1,3,2])

plt.scatter(x,y,color="orange")
```

```python
slope_curr=0
intercept_curr=0
learning_rate=0.01
epochs=1000

for i in range(epochs):
  y_predicted=slope_curr*x+intercept_curr

  sg=-(2/3)*sum(x*(y-y_predicted))
  sc=-(2/3)*sum(y-y_predicted)
  slope_curr=slope_curr-learning_rate*sg
  intercept_curr=intercept_curr-learning_rate*sc
  #plt.plot(x,y_predicted)
  
plt.scatter(x,y,color="orange")
plt.plot(x,y_predicted,color="hotpink")

  



```
